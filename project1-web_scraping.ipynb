# 2 Read .csv of multiples URLs with Pandas

<b> Link: roor + season + league </b>

```python
# link structure
"https://www.football-data.co.uk/mmz4281/"+"2223"+"/"+"E0"+".csv"
```

```python
# Create a variable root
root = "https://www.football-data.co.uk/mmz4281/"
```

# 2.1 Multiples leagues

- https://www.football-data.co.uk/mmz4281/2223/E0.csv
- https://www.football-data.co.uk/mmz4281/2223/E1.csv
- https://www.football-data.co.uk/mmz4281/2223/E2.csv
- https://www.football-data.co.uk/mmz4281/2223/E3.csv
- https://www.football-data.co.uk/mmz4281/2223/EC.csv


```python
# create leagues list
leagues = ['E0', 'E1', 'E2', 'E3', 'EC']
frames = []
for league in leagues:
    df = pd.read_csv(root+'2223'+'/'+league+'.csv', encoding='unicode_escape')
    frames.append(df)
    
# make a loop in leagues list, read multiples csv and attached
```

```python
# size of frames
len(frames)
```

```python
# show first, second and third frames
frames[4]
```

# 2.2 Multiple seasons

```python
# create leagues list
leagues = ['E0','E2', 'E3']
frames = []
for league in leagues:
    for season in range(15, 21):
        df = pd.read_csv(root+str(season)+str(season+1)+'/'+league+'.csv', encoding='unicode_escape')
        df.insert(1, 'season', season)
        frames.append(df)
    
# make a loop in leagues list, read multiples csv and attached
```

```python
# total frames (6 seasons x 3 leagues)
len(frames)
```

```python
# show first, second and third frames
frames[17]
```

# 3 Organize data in a dictionary

```python
# Create a dictionary with original name of the league
dict_countries = {
    'Spanish La Liga': 'SP1', 'Spanish Segunda Division':'SP2',
'German Bundesliga': 'D1',
'English Premier League': 'E0', 'English League 1': 'E2', 'English League 2':'E3'
}
```

```python
# get the element of dictionary
dict_countries['Spanish La Liga']
```

```python
for league in dict_countries:
    print(dict_countries[league])
```

```python
dict_historical_data = {}
# create leagues list
# leagues = ['E0','E2', 'E3']

for league in dict_countries:
    frames = []
    for season in range(15, 21):
        df = pd.read_csv(root+str(season)+str(season+1)+'/'+dict_countries[league]+'.csv')
        df.insert(1, 'season', season)
        frames.append(df)
    df_concat = pd.concat(frames)
    dict_historical_data[league]= df_concat
    
# make a loop in leagues list, read multiples csv and attached
```
